# **Top 10 batsmen based on past 3 years total runs scored**

```python
merged_1_3_df = pd.merge(df3, df1, on='match_id', how='inner')

merged_1_3_df['matchDate'] = pd.to_datetime(merged_1_3_df['matchDate'], format='%b %d, %Y', errors='coerce')

merged_1_3_df['Year'] = merged_1_3_df['matchDate'].dt.year
filtered_df = merged_1_3_df[merged_1_3_df['Year'].isin([2021, 2022, 2023])]

top_scorers = filtered_df.groupby('batsmanName')['runs'].sum().sort_values(ascending=False)
top_scorers.head(10)
```

# Top 10 batsmen based on past 3 years batting average. (min 60 balls faced in
# each season)

```python
balls_faced = filtered_df.groupby(['batsmanName', 'Year'])['balls'].sum().reset_index()
valid_batsmen = balls_faced[balls_faced['balls'] >= 60].groupby('batsmanName').count()
eligible_batsmen = valid_batsmen[valid_batsmen['Year'] ==  3].index

filtered_df = filtered_df[filtered_df['batsmanName'].isin(eligible_batsmen)]

batting_stats = filtered_df.groupby(['batsmanName']).agg(
    total_runs = ('runs', 'sum'),
    dismissals = ('out/not_out', lambda x: (x == 'out').sum())).reset_index()

batting_stats = batting_stats[batting_stats['dismissals'] > 0]

batting_stats['batting_average'] = batting_stats['total_runs'] / batting_stats['dismissals']

top_avg_batsmen = batting_stats.sort_values(by = 'batting_average', ascending=False).head(10)
top_avg_batsmen
```

#Top 10 batsmen based on past 3 years strike rate (min 60 balls faced in each
#season)

```python
strike_rate_stats = filtered_df.groupby('batsmanName').agg(
    total_runs = ('runs', 'sum'),
    total_balls = ('balls', 'sum')
).reset_index()

strike_rate_stats['strike_rate'] = (strike_rate_stats['total_runs']/strike_rate_stats['total_balls'])*100

top_strike_rate_batsmen = strike_rate_stats.sort_values(by='strike_rate', ascending=False).head(10)
top_strike_rate_batsmen
```

# **Top 10 bowlers based on past 3 years total wickets taken.**

```python
merged_1_4_df = pd.merge(df4, df1, on='match_id', how='inner')

merged_1_4_df['matchDate'] = pd.to_datetime(merged_1_3_df['matchDate'], format='%b %d, %Y', errors='coerce')

merged_1_4_df['Year'] = merged_1_4_df['matchDate'].dt.year
filtered_bowler_df = merged_1_4_df[merged_1_4_df['Year'].isin([2021, 2022, 2023])]

merged_1_4_df.groupby('bowlerName')['wickets'].sum().sort_values(ascending=False).head(10)
```

# **Top 10 bowlers based on past 3 years bowling average. (min 60 balls bowled in**
# **each season)**

```python
merged_1_4_df['bowlerBalls'] = merged_1_4_df['overs'] * 6

filtered_bowler_df = merged_1_4_df[merged_1_4_df['Year'].isin([2021, 2022, 2023])]

bowling_avg_stats = filtered_bowler_df.groupby('bowlerName').agg(
    total_runs_conducted = ('runs', 'sum'),
    total_wickets = ('wickets', 'sum'),
    total_bowler_balls = ('bowlerBalls', 'sum')
).reset_index()

eligible_bowlers = bowling_avg_stats[bowling_avg_stats['total_bowler_balls'] >= 600]

eligible_bowlers = eligible_bowlers[eligible_bowlers['total_wickets'] > 0]

eligible_bowlers['bowling_avg'] = eligible_bowlers['total_runs_conducted'] / eligible_bowlers['total_wickets']

top_avg_bowlers = eligible_bowlers.sort_values(by='bowling_avg').head(10)
top_avg_bowlers
```

#Top 10 bowlers based on past 3 years economy rate. (min 60 balls bowled in
#each season)


```python
bowling_economy_stats = filtered_bowler_df.groupby('bowlerName').agg(
    avg_economy = ('economy',  'mean'),
    total_bowler_balls = ('bowlerBalls', 'sum')
).reset_index()

eligible_economic_bowlers = bowling_economy_stats[bowling_economy_stats['total_bowler_balls'] >= 600]

top_economic_bowlers = eligible_economic_bowlers.sort_values(by='avg_economy').head(10)
top_economic_bowlers
```

# **Top 5 batsmen based on past 3 years boundary % (fours and sixes).**

```python
boundary_stats = filtered_df.groupby('batsmanName').agg(
    total_fours = ('4s', 'sum'),
    total_sixes = ('6s', 'sum'),
    total_balls = ('balls', 'sum')
).reset_index()

boundary_stats['boundary_percentage'] = ((boundary_stats['total_fours'] + boundary_stats['total_sixes']) / boundary_stats['total_balls']) * 100

top_boundary_batsmen = boundary_stats.sort_values(by = 'boundary_percentage', ascending=False).head(5)
top_boundary_batsmen
```

# **Top 5 bowlers based on past 3 years dot ball %**

```python
dot_ball_stats = filtered_bowler_df.groupby('bowlerName').agg(
    total_dots = ('0s', 'sum'),
    total_balls = ('bowlerBalls', 'sum')
).reset_index()

dot_ball_stats = dot_ball_stats[dot_ball_stats['total_balls'] >= 600]

dot_ball_stats['dot_ball_percentage'] = (dot_ball_stats['total_dots'] / dot_ball_stats['total_balls']) * 100

top_dot_bowlers = dot_ball_stats.sort_values(by = 'dot_ball_percentage', ascending=False).head(5)
top_dot_bowlers
```

# **Top 4 teams based on past 3 years winning %.**

```python
team_matches = df1['team1'].value_counts() + df1['team2'].value_counts()

team_wins = df1['winner'].value_counts()

winning_percentage = (team_wins / team_matches) * 100

top_winning_teams = winning_percentage.sort_values(ascending=False).head(4)
top_winning_teams
```

#Top 2 teams with the highest number of wins achieved by chasing targets over
